## Aggregation (Groupby)

```python
df['Brand'].value_counts()
```

```python
brand_df = df.groupby(['Brand']).size().reset_index(name='counts')
brand_df
```

```python
df.groupby(['category']) \
  .agg({'n_words':'mean'}) \
  .sort_values(by='n_words', ascending=False) \
  .plot(kind='bar', figsize=(8,4))
```
